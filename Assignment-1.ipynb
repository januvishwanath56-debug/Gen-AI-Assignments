```python
# Task 2:Build Advanced Preprocessing Function

import re
from collections import Counter
EXCEPTIONS = {"no", "not"}

def preprocess_text(text):
    if not isinstance(text, str) or not text.strip():
        return [], ""

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|\S+@\S+", " ", text)

    # Remove numbers
    text = re.sub(r"\d+", " ", text)

    # Remove emojis & non-ASCII characters
    text = text.encode("ascii", "ignore").decode()

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove punctuation
    text = re.sub(r"[^\w\s]", " ", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenize
    tokens = text.split()

    # Remove very short tokens except meaningful ones
    tokens = [t for t in tokens if len(t) > 2 or t in EXCEPTIONS]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

```python
# Task 3: Stress Testing

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []

for s in test_sentences:
    tokens, clean = preprocess_text(s)
    all_tokens.extend(tokens)

    print("Original:", s)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print()
```

```python
# Task 4: Token Analytics

for s in test_sentences:
    tokens, _ = preprocess_text(s)

    if tokens:
        avg_len = sum(len(t) for t in tokens) / len(tokens)
    else:
        avg_len = 0

    print("Sentence:", s)
    print("Total tokens:", len(tokens))
    print("Unique tokens:", len(set(tokens)))
    print("Average token length:", round(avg_len, 2))
    print()
```

```python
# Task 5: Frequency Analysis

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[:-6:-1])
```

```python
# Task 6: Build Full Pipeline

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

pipeline_output = full_pipeline(test_sentences)
print(pipeline_output)
```

```python
# Task 7: Error Handling

edge_cases = ["", "😍😍😍", "1234567890"]

for case in edge_cases:
    tokens, clean = preprocess_text(case)

    print("Input:", repr(case))
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-" * 40)
```
